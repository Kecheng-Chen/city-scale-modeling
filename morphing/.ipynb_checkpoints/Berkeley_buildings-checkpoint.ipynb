```python
import geojson
from shapely import geometry as geo
from shapely.ops import transform as trans
from matplotlib import pyplot as plt
import pandas as pd
import utm
import numpy as np
import csv
from energy_read import length_generate
```

```python
# Get the street blocks (data from OSM)
with open('berkeley2.geojson') as f:
    gj = geojson.load(f)
features = gj['features'][1]
list_polygon = []
for f in gj['features']:
    poly = geo.shape(f['geometry'])
    list_polygon.append(poly)
```

```python
# Plot the street blocks
plt.figure(1)
for i in range(0, len(list_polygon)):
    mpoly = list_polygon[i]
    #dmd = list_blkdmdbhlenth[i]
    if mpoly.geom_type == 'MultiPolygon':
        print("Multi poly")
        for polygon in mpoly:
            px,py = polygon.exterior.xy
            plt.plot(px,py, 'k--', linewidth=0.3)
            #plt.fill(x,y, color=cmap(norm(dmd)))
    elif mpoly.geom_type == 'Polygon':
        px,py = mpoly.exterior.xy
        plt.plot(px,py, 'k--', linewidth=0.1)
        #plt.fill(x,y, color=cmap(norm(dmd)))
plt.savefig('street_blocks.png',dpi=1200)
```

```python
building_data = pd.read_csv('Berkeley_UrbanSim.csv')
lat = building_data['Latitude'].to_numpy()
long = building_data['Longitude'].to_numpy()
tid = building_data['Type ID'].to_numpy()
barea = building_data['Area'].to_numpy()
byear = building_data['Year Built'].to_numpy().astype(int)
x = np.zeros(len(lat))
y = np.zeros(len(lat))
for i in range(0, lat.shape[0]):
    u = utm.from_latlon(lat[i], long[i])
    x[i] = u[0]
    y[i] = u[1]
btype = []
cm = []
for t in tid:
    if t == 0 or t==15 or t==16:
        btype.append('Other/Unknown')
        cm.append(0)
    elif t==1 or t==2 or t==3 or t==12:
        btype.append('Residential')
        cm.append(1)
    elif t==4 or t==14:
        btype.append('Office')
        cm.append(2)
    elif t==5:
        btype.append('Hotel')
        cm.append(3)
    elif t==6:
        btype.append('School')
        cm.append(4)
    elif t==7 or t==8 or t==9:
        btype.append('Industrial')
        cm.append(5)
    elif t==10 or t==11 or t==13:
        btype.append('Retail')
        cm.append(6)
    #elif t==15 or t==16:
    #    btype.append('Other/Unknown')#parking
    #    cm.append(0)
    else:
        print("Unidentified type ID")

#btypelabels=['other', 'residential','office','hotel','school','industrial','retail','parking']
btypelabels=['Other/Unknown', 'Residential','Office','Hotel','School','Industrial','Retail']
```

```python
plt.figure(2)
#plt.plot(x,y,'x', markersize=0.1)
ss = plt.scatter(x,y,s=0.02,marker='x',c=cm)
lgd = plt.legend(handles=ss.legend_elements()[0],labels=btypelabels,bbox_to_anchor=(1,1))
plt.savefig('building_points.png',dpi=1200,bbox_extra_artists=(lgd,),bbox_inches='tight')
```

```python
# Get the building footprints (data from chaofeng)
def latlong_to_utm(lat, long, z=None):
    u = utm.from_latlon(long,lat)
    return u[0], u[1]

with open('Berkeley_BIM.geojson') as f:
    fgj = geojson.load(f)
features = fgj['features'][1]
list_fp_polygon = []
for f in fgj['features']:
    fppoly = geo.shape(f['geometry'])
    transpoly = trans(latlong_to_utm, fppoly)
    list_fp_polygon.append(transpoly)
```

```python
# Plot the street blocks
plt.figure(4)
for i in range(0, len(list_fp_polygon)):
    mpoly = list_fp_polygon[i]
    px,py = mpoly.exterior.xy
    plt.plot(px,py, 'b-', linewidth=0.1)
    #plt.fill(px,py, color='b')
plt.savefig('footprints.png',dpi=1200)
```

```python
# Plot the street blocks
plt.figure(5)
for i in range(0, len(list_polygon)):
    mpoly = list_polygon[i]
    #dmd = list_blkdmdbhlenth[i]
    if mpoly.geom_type == 'MultiPolygon':
        print("Multi poly")
        for polygon in mpoly:
            px,py = polygon.exterior.xy
            plt.plot(px,py, 'k--', linewidth=0.3)
            #plt.fill(x,y, color=cmap(norm(dmd)))
    elif mpoly.geom_type == 'Polygon':
        px,py = mpoly.exterior.xy
        plt.plot(px,py, 'k--', linewidth=0.1)
        #plt.fill(x,y, color=cmap(norm(dmd)))
for i in range(0, len(list_fp_polygon)):
    mpoly = list_fp_polygon[i]
    px,py = mpoly.exterior.xy
    px_utm = []
    py_utm = []
    for j in range(0,len(px)):
        u = utm.from_latlon(py[j],px[j])
        px_utm.append(u[0])
        py_utm.append(u[1])
    plt.plot(px_utm,py_utm, 'b-', linewidth=0.1)
plt.savefig('block_footprints.png',dpi=1600)
```

```python
list_nonempty_polygon_id = []
list_polygon_point_id = []
for i in range(0, len(list_polygon)):
    polygon_point_id = []
    for j in range(0, len(x)):
        pp = geo.Point(x[j], y[j])
        if list_polygon[i].contains(pp):
            polygon_point_id.append(j)
    if len(polygon_point_id) != 0:
        list_nonempty_polygon_id.append(i)
        list_polygon_point_id.append(polygon_point_id)
```

```python
df = pd.DataFrame(list_nonempty_polygon_id, columns=["Poly_id"])
df.to_csv('nonempty_polygon_id.csv', index=False)
df2 = pd.DataFrame(list_polygon_point_id)
df2.to_csv('polygon_point_id.csv', index=False,header=False)
```

```python
list_polygon_point_id = []
with open('Match_records/polygon_point_id.csv', newline='') as f:
    reader=csv.reader(f)
    list_polygon_point_id = list(reader)
for i in range(0, len(list_polygon_point_id)):
    list_polygon_point_id[i] = [ll for ll in list_polygon_point_id[i] if ll != '']
    for j in range(0, len(list_polygon_point_id[i])):
        list_polygon_point_id[i][j] = int(float(list_polygon_point_id[i][j]))
        
list_nonempty_polygon_id = []
with open('Match_records/nonempty_polygon_id.csv', newline='') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        list_nonempty_polygon_id.append(int(row[0]))
```

```python
plt.figure(6)
x_insb = []
y_insb = []
cm_insb = []
for polyp_list in list_polygon_point_id:
    for polyp in polyp_list:
        x_insb.append(x[polyp])
        y_insb.append(y[polyp])
        cm_insb.append(cm[polyp])
ss = plt.scatter(x_insb,y_insb,s=0.02,marker='x',c=cm_insb)
lgd = plt.legend(handles=ss.legend_elements()[0],labels=btypelabels,bbox_to_anchor=(1,1))
for i in range(0, len(list_nonempty_polygon_id)):
    mpoly = list_polygon[list_nonempty_polygon_id[i]]
    if mpoly.geom_type == 'MultiPolygon':
        print("Multi poly")
        for polygon in mpoly:
            px,py = polygon.exterior.xy
            plt.plot(px,py, 'k--', linewidth=0.3)
    elif mpoly.geom_type == 'Polygon':
        px,py = mpoly.exterior.xy
        plt.plot(px,py, 'k--', linewidth=0.1)
plt.savefig('matched_block_points.png',dpi=1600,bbox_extra_artists=(lgd,),bbox_inches='tight')
```

```python
list_polygon_footprint_id = []
for i in range(0, len(list_nonempty_polygon_id)):
    polygon_footprint_id = []
    for j in range(0, len(list_fp_polygon)):
        pp = list_fp_polygon[j].centroid
        if list_polygon[list_nonempty_polygon_id[i]].contains(pp):
            polygon_footprint_id.append(j)
    list_polygon_footprint_id.append(polygon_footprint_id)
```

```python
df3 = pd.DataFrame(list_polygon_footprint_id)
df3.to_csv('polygon_footprint_id.csv', index=False,header=False)
```

```python
# Plot the street blocks
plt.figure(7)
for i in range(0, len(list_nonempty_polygon_id)):
    mpoly = list_polygon[list_nonempty_polygon_id[i]]
    #dmd = list_blkdmdbhlenth[i]
    px,py = mpoly.exterior.xy
    plt.plot(px,py, 'k--', linewidth=0.1)
    #plt.fill(x,y, color=cmap(norm(dmd)))
    for j in range(0,len(list_polygon_footprint_id[i])):
        fp_poly = list_fp_polygon[list_polygon_footprint_id[i][j]]
        fpx,fpy = fp_poly.exterior.xy
        plt.plot(fpx,fpy, 'b-', linewidth=0.1)
plt.savefig('matched_block_footprints.png',dpi=1600)
```

```python
list_polygon_footprint_area = []
list_polygon_footprint_coverage = []
for polygon_footprint_ids in list_polygon_footprint_id:
    area = 0
    for fpid in polygon_footprint_ids:
        area = area + list_fp_polygon[fpid].area
    list_polygon_footprint_area.append(area)
for i in range(0, len(list_nonempty_polygon_id)):
    poly_area = list_polygon[list_nonempty_polygon_id[i]].area
    coverage = list_polygon_footprint_area[i] / poly_area * 100
    list_polygon_footprint_coverage.append(coverage)
```

```python
print(min(list_polygon_footprint_coverage))
```

```python
#storage length
plt.figure(8)
cmap = plt.cm.RdYlBu
norm = plt.Normalize(0, 100)
#norm = plt.Normalize(min(list_blkdmdbhlenth), 0.02*max(list_blkdmdbhlenth))
for i in range(0, len(list_nonempty_polygon_id)):
    mpoly = list_polygon[list_nonempty_polygon_id[i]]
    cover = list_polygon_footprint_coverage[i]
    px, py = mpoly.exterior.xy
    plt.fill(px,py, color=cmap(norm(cover)))
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array(list_polygon_footprint_coverage)
plt.colorbar(sm)
plt.savefig('block_footprint_coverage.png',dpi=1600)
```

```python
list_polygon_bhlgh = []
for i in range(0, len(list_nonempty_polygon_id)):
    tmp_point_ids = list_polygon_point_id[i]
    tmp_list_barea = []
    tmp_list_byear = []
    tmp_list_btype = []
    for pid in tmp_point_ids:
        tmp_list_barea.append(barea[pid])
        tmp_list_byear.append(byear[pid])
        tmp_list_btype.append(btype[pid])
    #print(tmp_list_barea)
    #print(tmp_list_byear)
    #print(tmp_list_btype)
    sum_tarea = 0
    for i in range(0, len(tmp_list_barea)):
        if tmp_list_btype[i] != 'Other/Unknown':
            sum_tarea = sum_tarea + tmp_list_barea[i]
    if len(tmp_list_barea) == 0 or sum_tarea == 0:
        list_polygon_bhlgh.append(0)
    else:
        list_polygon_bhlgh.append(length_generate(tmp_list_btype,tmp_list_barea,tmp_list_byear))
    #print(list_polygon_bhlgh)
```
