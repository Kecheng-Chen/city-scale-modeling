```python
# Utility functions
def moving_average(a, n=3):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    med = ret[n-1:] / n
    nedge = math.floor(n/2)
    
    ma = np.zeros((a.shape))
    for i in range(0,nedge):
        ma[i] = np.sum(a[0:(i+nedge+1)]) / (i+1+nedge)
        ma[-1-i] = np.sum(a[-i-nedge-1:]) / (i+1+nedge)
    
    ma[nedge:(nedge+len(med))] = med
    
    return ma

def mean_confidence_interval(data, confidence = 0.9):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), stats.sem(a)
    h = se * stats.t.ppf((1+confidence)/2., n-1)
    return m, m-h, m+h

def get_design_parameters(f_in, month_disp):
    # Number of data input
    ndata = f_in.shape[1]
    
    # Part load factor
    plf = np.ndarray(shape=(1,ndata))
    
    # Peak heating and cooling load
    mheat = f_in.min(axis=0)
    mcool = f_in.max(axis=0)
    
    # Net annual load
    net = np.ndarray(shape=(1,ndata))
    
    
    
    print(np.sum(t, axis=0))
print(t.max(axis=0))
print(t.min(axis=0))
```

```python
# Parameters
# Number of random annual curve to generate
ncurve = 20
# Number of daily curves in one annual curve
nday = 365
# Number of sample points in one daily curve
nhour = 24
```

```python
# Read data from CSV files
data_0 = np.genfromtxt('hourly_demand/RESIDENT.csv', delimiter=',')
data_list = []
ndata = 20
for i in range(0,ndata):
    filename = 'auunal_curve_20/' + str(i+1) + '.csv'
    data_tmp = np.genfromtxt(filename, delimiter=',')
    data_list.append(data_tmp)
```

```python
# Method 1, sampling score month by month
# do fpca together? or month by month?

month_disp = [0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]

m1_f_0 = data_0[1:None,1]
m1_f_daily = np.transpose(data_0[1:None,1].reshape(nday,nhour))
time = np.arange(1,nday+1)
time = np.float64(time)
```

```python
# fPCA on f
m1_fcg_fnorm = fcg(m1_f_daily)
m1_fcg_fnorm.calc_fpca(False, True, 100, 20, 20)
m1_fcg_fnorm.distribution_est_piecewise('gaussian', month_disp)
```

```python
m1_f_annual = np.zeros((nday*nhour, ncurve))

for i in range(0, ncurve):
    gen_f = m1_fcg_fnorm.curve_gen_piecewise(month_disp)
    m1_f_annual[:,i] = np.transpose(gen_f).reshape(-1)
```

```python
# Method 2, for single annual curve, extract median and std, then do fpca
m2_f_0 = data_0[1:None,1]
m2_f_daily = np.transpose(data_0[1:None,1].reshape(nday,nhour))
time = np.arange(1,nday+1)
time = np.float64(time)

# Record med and std
record_med = np.zeros((nday, 1))
record_std = np.zeros((nday, 1))
for i in range(0, nday):
    record_med[i] = np.median(m2_f_daily[:,i])
    record_std[i] = np.std(m2_f_daily[:,i])
# Normalize f using med and std
m2_f_normalized = np.zeros((nhour, nday))
for i in range(0, nday):
    m2_f_normalized[:,i] = (m2_f_daily[:,i] - record_med[i,0])/record_std[i,0]
```

```python
# fPCA on normalized f
m2_fcg_fnorm = fcg(m2_f_normalized)
m2_fcg_fnorm.calc_fpca(False, True, 100, 20, 20)
m2_fcg_fnorm.distribution_est('gaussian')
```

```python
m2_f_annual = np.zeros((nday*nhour, ncurve))

for i in range(0, ncurve):
    gen_f = m2_fcg_fnorm.curve_gen(nday)
    for j in range(0, nday):
        gen_f[:,j] = gen_f[:,j]*record_std[j] + record_med[j]
    gen_f_array = np.transpose(gen_f).reshape(-1)
    m2_f_annual[:,i] = gen_f_array
```

```python
# Method 3, for multiple annual curves, extract median and std
# then do fpca on median, std and normalized daily curves respectively
n_indata = 21
n_mv = 15
m3_f_list = []
m3_f_daily = np.zeros((nhour, nday*n_indata))

m3_f_list.append(data_0[1:None,1])
for i in range(0, ndata):
    m3_f_list.append(data_list[i][1:None,1])
    
m3_f_daily[:,0:365] = np.transpose(data_0[1:None,1].reshape(nday,nhour))
for i in range(0, ndata):
    m3_f_daily[:, nday*(i+1):nday*(i+2)] = np.transpose(data_list[i][1:None,1].reshape(nday,nhour))
    
# Record med and std
record_med = np.zeros((nday, n_indata))
record_std = np.zeros((nday, n_indata))
#record_med_std = np.zeros((2*nday, n_indata))
for i in range(0, n_indata):
    for j in range(0, nday):
        record_med[j,i] = np.median(m3_f_daily[:,i*nday+j])
        record_std[j,i] = np.std(m3_f_daily[:,i*nday+j])
        #record_med_std[j,i] = np.median(m3_f_daily[:,i*nday+j])
        #record_med_std[j+nday,i] = np.std(m3_f_daily[:,i*nday+j])

plt.figure()
for i in range(0,21):
    plt.plot(np.arange(1,nday+1), record_med[:,i])
plt.figure()
for i in range(0,21):
    plt.plot(np.arange(1,nday+1), record_std[:,i])

# Smooth data and take ln for std
for i in range(0, n_indata):
    record_med[:,i] = moving_average(record_med[:,i], n_mv)
    record_std[:,i] = moving_average(record_std[:,i], n_mv)

# Normalize f using smoothed med and std
m3_f_normalized = np.zeros((nhour, nday*n_indata))
for i in range(0, nday*n_indata):
    row = i%nday
    col = math.floor(i/nday)
    m3_f_normalized[:,i] = (m3_f_daily[:,i] - record_med[row,col])/record_std[row,col]

record_std = np.log(record_std)

plt.figure()
for i in range(0,21):
    plt.plot(np.arange(1,nday+1), record_med[:,i])
plt.figure()
for i in range(0,21):
    plt.plot(np.arange(1,nday+1), record_std[:,i])
#plt.figure()
#for i in range(0,21):
#    plt.plot(np.arange(1,2*nday+1), record_med_std[:,i])
```
